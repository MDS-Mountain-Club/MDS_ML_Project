```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})


# Original Series
fig, axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(solar); axes[0, 0].set_title('Original Series')
plot_acf(solar, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(solar.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(solar.diff(), ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(solar.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(solar.diff().diff(), ax=axes[2, 1])

plt.show()
```

```python
# 1,1,2 ARIMA Model
model = ARIMA(solar, order=(1,1,2))
model_fit = model.fit()
print(model_fit.summary())
```

```python
residuals = pd.DataFrame(model_fit.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```

```python
model.plot_diagnostics(figsize=(10,8))
plt.show()
```

```python
# Actual vs Fitted
fig, ax = plt.subplots()
ax = solar.plot(ax=ax)
plot_predict(model_fit, ax=ax, alpha=0.05, dynamic=False)
plt.show()
```

```python
forecast = model_fit.forecast(steps=10)
print(forecast)
```

```python
plt.plot(solar, label='Original Data')

plt.plot(pd.date_range(start=solar, periods=10, freq='M'), forecast, label='Forecast', color='red')
plt.title('ARIMA Model Forecast')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
