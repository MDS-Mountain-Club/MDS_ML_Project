```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.kernel_ridge import KernelRidge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
df_final = pd.read_csv('../Data/df_final.csv', index_col="time")
```

```python
X = df_final.drop(columns=['generation solar','generation wind onshore', 'forecast wind onshore day ahead', 'forecast solar day ahead'])
y_solar = df_final['generation solar']
y_wind = df_final['generation wind onshore']
```

```python
def split_by_date(X,train_end_date):
    if not isinstance(train_end_date, pd.Timestamp):
        train_end_date = pd.Timestamp(train_end_date)

    # Convert index of X and y to Timestamp objects if they are strings
    if isinstance(X.index[0], str):
        X.index = pd.to_datetime(X.index)
        
    X_train = X[X.index <= train_end_date]
    X_test = X[X.index > train_end_date]
    
    return X_train, X_test

train_end_date = pd.to_datetime('2017-12-31 23:00:00+00:00')
X_train, X_test = split_by_date(X,train_end_date)
y_solar_train,y_solar_test = split_by_date(y_solar,train_end_date)

# Standardize all columns except target
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_train = pd.DataFrame(X_train_scaled, index=X_train.index, columns=X_train.columns)

X_test_scaled = scaler.transform(X_test)
X_test = pd.DataFrame(X_test_scaled, index=X_test.index, columns=X_test.columns)
```

```python
# Kernel Ridge Regression with Linear kernel
kr_solar_linear = KernelRidge(kernel='linear')
kr_solar_linear.fit(X_train, y_solar_train)
y_pred_solar_linear = kr_solar_linear.predict(X_test)
mse_linear = mean_squared_error(y_solar_test, y_pred_solar_linear)
print("MSE for Linear Kernel:", mse_linear)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(y_solar_test.index, y_solar_test, label='Actual Solar Production', color='blue')
plt.plot(y_solar_test.index, y_pred_solar_linear, label='Predicted Solar Production (Linear)', linestyle='--', color='green')
plt.title('Solar Production Prediction')
plt.xlabel('Time')
plt.ylabel('Solar Production')
plt.legend()
plt.show()
```

```python
# Kernel Ridge Regression with RBF kernel
kr_rbf = KernelRidge(kernel='rbf')
parameters = {'alpha': [0.1, 0.5], 'gamma': [0.02, 0.04]}
grid_search_solar_rbf = GridSearchCV(kr_rbf, parameters, cv=5, scoring='neg_mean_squared_error')
grid_search_solar_rbf.fit(X_train, y_solar_train)

best_params_solar_rbf = grid_search_solar_rbf.best_params_
best_model_solar_rbf = grid_search_solar_rbf.best_estimator_
```

```python
print("Best Parameters for RBF Kernel:", best_params_solar_rbf)

# Train the best RBF model
best_model_solar_rbf.fit(X_train, y_solar_train)

# Predictions
y_pred_solar_rbf = best_model_solar_rbf.predict(X_test)

# Calculate MSE
mse_solar_rbf = mean_squared_error(y_solar_test, y_pred_solar_rbf)
print("MSE for RBF Kernel:", mse_solar_rbf)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(y_solar_test.index, y_solar_test, label='Actual Solar Production', color='blue')
plt.plot(y_solar_test.index, y_pred_solar_rbf, label='Predicted Solar Production (RBF)', linestyle='--', color='orange')
plt.title('Solar Production Prediction')
plt.xlabel('Time')
plt.ylabel('Solar Production')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.kernel_ridge import KernelRidge
from sklearn.metrics import mean_squared_error
from sklearn.metrics.pairwise import pairwise_kernels  # Add this import

# Define the hyperparameter grid
parameters = {'alpha': [0.1, 0.5], 'gamma': [0.02, 0.04]}

def periodic(X, Y=None):
    if Y is None:
        Y = X
    # Reshape input data to ensure it has 2 dimensions
    X = np.atleast_2d(X)
    Y = np.atleast_2d(Y)
    pairwise_differences = pairwise_kernels(X, Y, metric=lambda x, y: np.abs((x - y)))
    return np.exp(-parameters['gamma'] * pairwise_differences)

# Kernel Ridge Regression with PER kernel
kr_per = KernelRidge(kernel=periodic)

# Perform grid search
grid_search_solar_per = GridSearchCV(kr_per, parameters, cv=5, scoring='neg_mean_squared_error')
grid_search_solar_per.fit(X_train, y_solar_train)

# Get the best parameters
best_params_solar_per = grid_search_solar_per.best_params_
best_model_solar_per = grid_search_solar_per.best_estimator_

print("Best Parameters for PER Kernel:", best_params_solar_per)

# Train the best PER model
best_model_solar_per.fit(X_train, y_solar_train)

# Predictions
y_pred_solar_per = best_model_solar_per.predict(X_test)

# Calculate MSE
mse_solar_per = mean_squared_error(y_solar_test, y_pred_solar_per)
print("MSE for PER Kernel:", mse_solar_per)
```
